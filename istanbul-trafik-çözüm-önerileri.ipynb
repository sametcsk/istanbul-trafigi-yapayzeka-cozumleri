```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, BatchNormalization, GRU
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l1_l2
try:
    import skfuzzy as fuzz
    from skfuzzy import control as ctrl
except ImportError:
    print(" scikit-fuzzy yükleniyor...")
    !pip install -q scikit-fuzzy
    import skfuzzy as fuzz
    from skfuzzy import control as ctrl
import warnings
warnings.filterwarnings('ignore')


plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")




from google.colab import drive
drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/'
DOSYALAR = [
    'eylül24_trafik.csv', 'ekim24_trafik.csv', 'kasım24_trafik.csv',
    'aralık24_trafik.csv', 'ocak25_trafik.csv'
]

veri_listesi = []
for dosya in DOSYALAR:
    try:
        df_temp = pd.read_csv(BASE_PATH + dosya, encoding='utf-8')
        veri_listesi.append(df_temp)
        print(f"   ✓ {dosya:25s} → {len(df_temp):>9,} satır")
    except Exception as e:
        print(f"   ✗ {dosya:25s} → {str(e)[:50]}")

ana_veri = pd.concat(veri_listesi, ignore_index=True)
print(f"\n✅ Toplam: {len(ana_veri):,} satır")


print("\n" + "="*80)
print("FEATURE ENGINEERING")
print("="*80)

ana_veri['DATE_TIME'] = pd.to_datetime(ana_veri['DATE_TIME'], errors='coerce')
ana_veri.dropna(subset=['DATE_TIME', 'AVERAGE_SPEED'], inplace=True)

print("\n Aykırı değer temizleme")
Q1 = ana_veri['AVERAGE_SPEED'].quantile(0.01)
Q3 = ana_veri['AVERAGE_SPEED'].quantile(0.99)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR
onceki_boyut = len(ana_veri)
ana_veri = ana_veri[(ana_veri['AVERAGE_SPEED'] >= alt_sinir) &
                    (ana_veri['AVERAGE_SPEED'] <= ust_sinir)]
print(f"   ✓ {onceki_boyut - len(ana_veri):,} aykırı değer temizlendi")

# Temel zaman özellikleri
ana_veri['Yil'] = ana_veri['DATE_TIME'].dt.year
ana_veri['Ay'] = ana_veri['DATE_TIME'].dt.month
ana_veri['Gun'] = ana_veri['DATE_TIME'].dt.day
ana_veri['Saat'] = ana_veri['DATE_TIME'].dt.hour
ana_veri['DakikaGunun'] = ana_veri['DATE_TIME'].dt.hour * 60 + ana_veri['DATE_TIME'].dt.minute
ana_veri['Haftanin_Gunu'] = ana_veri['DATE_TIME'].dt.dayofweek
ana_veri['HaftaSonu'] = (ana_veri['Haftanin_Gunu'] >= 5).astype(int)

# Periyodik özellikler
ana_veri['Saat_Sin'] = np.sin(2 * np.pi * ana_veri['Saat'] / 24)
ana_veri['Saat_Cos'] = np.cos(2 * np.pi * ana_veri['Saat'] / 24)
ana_veri['Ay_Sin'] = np.sin(2 * np.pi * ana_veri['Ay'] / 12)
ana_veri['Ay_Cos'] = np.cos(2 * np.pi * ana_veri['Ay'] / 12)
ana_veri['Gun_Hafta_Sin'] = np.sin(2 * np.pi * ana_veri['Haftanin_Gunu'] / 7)
ana_veri['Gun_Hafta_Cos'] = np.cos(2 * np.pi * ana_veri['Haftanin_Gunu'] / 7)

# Zaman dilimleri
def zaman_dilimi_func(saat):
    if 6 <= saat < 10: return 'Sabah_Yogunluk'
    elif 10 <= saat < 17: return 'Gunduz'
    elif 17 <= saat < 21: return 'Aksam_Yogunluk'
    else: return 'Gece'

ana_veri['Zaman_Dilimi'] = ana_veri['Saat'].apply(zaman_dilimi_func)

# Tatiller
tatil_gunleri = pd.to_datetime(['2024-09-29', '2024-10-29', '2024-12-31', '2025-01-01'])
ana_veri['Tatil_Gunu'] = ana_veri['DATE_TIME'].dt.date.isin(tatil_gunleri.date).astype(int)

# Lokasyon özellikleri
geohash_stats = ana_veri.groupby('GEOHASH')['AVERAGE_SPEED'].agg(['mean', 'std', 'min', 'max'])
ana_veri['Lokasyon_Ort_Hiz'] = ana_veri['GEOHASH'].map(geohash_stats['mean'])
ana_veri['Lokasyon_Std_Hiz'] = ana_veri['GEOHASH'].map(geohash_stats['std'])

# Sıralama
ana_veri = ana_veri.sort_values(['GEOHASH', 'DATE_TIME']).reset_index(drop=True)

print(f"✅ Özellik sayısı: {ana_veri.shape[1]}")


print("\n" + "="*80)
print("KARA NOKTA TESPİTİ")
print("="*80)

yogun_saatler_df = ana_veri[
    ((ana_veri['Saat'] >= 7) & (ana_veri['Saat'] < 10)) |
    ((ana_veri['Saat'] >= 17) & (ana_veri['Saat'] < 20))
].copy()

print("\n" + "="*80)
print("STRATEJİK ANALİZ - KARA NOKTA TESPİTİ")
print("="*80)

print("\n Altyapı Yatırım Öncelikleri")
# Yoğun saatlerde sıkışıklık endeksi
yogun_saatler_df = ana_veri[
    ((ana_veri['Saat'] >= 7) & (ana_veri['Saat'] < 10)) |
    ((ana_veri['Saat'] >= 17) & (ana_veri['Saat'] < 20))
].copy()

yogun_saatler_df['Sıkısıklık_Endeksi'] = (
    yogun_saatler_df['NUMBER_OF_VEHICLES'] / (yogun_saatler_df['AVERAGE_SPEED'] + 1)
)

kara_noktalar = yogun_saatler_df.groupby('GEOHASH').agg(
    Ort_Sıkısıklık=('Sıkısıklık_Endeksi', 'mean'),
    Ort_Hiz=('AVERAGE_SPEED', 'mean'),
    Ort_Arac=('NUMBER_OF_VEHICLES', 'mean')
).nlargest(10, 'Ort_Sıkısıklık').reset_index()

print("\n" + "-"*80)
print("     İSTANBUL'UN EN KRİTİK 10 TRAFİK KARA NOKTASI")
print("-"*80)
for idx, row in kara_noktalar.iterrows():
    print(f"   {idx+1:2d}. {row['GEOHASH']:12s} | Sıkışıklık: {row['Ort_Sıkısıklık']:6.2f} | "
          f"Hız: {row['Ort_Hiz']:5.1f} km/h | Araç: {row['Ort_Arac']:6.0f}")
print("-"*80)

print("\n ÖNERİ (Stratejik Planlama):")
print("   • Bu lokasyonlara yapılacak köprülü kavşak, alt geçit, ek şerit gibi altyapı yatırımları önceliklendirilmelidir.")


print("\n" + "="*80)
print(" VERİ OPTİMİZASYONU")
print("="*80)

# Saatlik ortalama
ana_veri['Tarih'] = ana_veri['DATE_TIME'].dt.date
ana_veri['Saat_Tam'] = ana_veri['DATE_TIME'].dt.hour

agg_funcs = {
    'AVERAGE_SPEED': 'mean',
    'NUMBER_OF_VEHICLES': 'mean',
    'LATITUDE': 'first',
    'LONGITUDE': 'first',
    'Haftanin_Gunu': 'first',
    'HaftaSonu': 'first',
    'Ay': 'first',
    'Tatil_Gunu': 'first',
    'Lokasyon_Ort_Hiz': 'first',
    'Lokasyon_Std_Hiz': 'first',
    'Saat_Sin': 'first',
    'Saat_Cos': 'first',
    'Ay_Sin': 'first',
    'Ay_Cos': 'first',
    'Gun_Hafta_Sin': 'first',
    'Gun_Hafta_Cos': 'first',
    'DakikaGunun': 'first',
    'Zaman_Dilimi': lambda x: x.mode()[0] if len(x.mode()) > 0 else x.iloc[0]
}

veri_saatlik = ana_veri.groupby(['GEOHASH', 'Tarih', 'Saat_Tam']).agg(agg_funcs).reset_index()
veri_saatlik['DATE_TIME'] = pd.to_datetime(veri_saatlik['Tarih']) + pd.to_timedelta(veri_saatlik['Saat_Tam'], unit='h')
veri_saatlik.drop(['Tarih', 'Saat_Tam'], axis=1, inplace=True)
veri_saatlik['Saat'] = veri_saatlik['DATE_TIME'].dt.hour

# Top 50 yoğun lokasyon
top_lokasyonlar = ana_veri.groupby('GEOHASH')['NUMBER_OF_VEHICLES'].sum().nlargest(50).index
df = veri_saatlik[veri_saatlik['GEOHASH'].isin(top_lokasyonlar)].copy()

print(f"   ✓ Saatlik veri: {len(veri_saatlik):,} satır")
print(f"   ✓ Final veri: {len(df):,} satır")

print("\n" + "="*80)
print(" MOBİL SİNYAL VERİSİ ENTEGRASYONU (SİMÜLASYON)")
print("="*80)

# df (veri_saatlik, top 50 lokasyon).
df["Sinyal_Yogunlugu"]=df["NUMBER_OF_VEHICLES"]*1.3 #her araçta ortalama 1.3 kişi
saat_etkisi={
    12: 1.3 ,13: 1.3,
    18: 1.2 ,19: 1.2 ,20:1.2,

}

df["Sinyal_Yogunlugu"]=df.apply(lambda row: row["Sinyal_Yogunlugu"]*saat_etkisi.get(row["Saat"],1),axis=1)

df["Sinyal_Yogunlugu"]=np.where(
    (df["HaftaSonu"]==1)&(df["Saat"]>=12)&(df["Saat"]<=22),
    df["Sinyal_Yogunlugu"]*1.3,
    df["Sinyal_Yogunlugu"]
 )
noise=np.random.normal(1,0.05,df.shape[0])
df["Sinyal_Yogunlugu"]=(df["Sinyal_Yogunlugu"]*noise).astype(int)

print("✅ 'Sinyal_Yogunlugu' adında yeni bir özellik başarıyla oluşturuldu.")
print("\nÖrnek Sinyal Yoğunluğu verisi:")
print(df[['DATE_TIME', 'GEOHASH', 'NUMBER_OF_VEHICLES', 'Sinyal_Yogunlugu']].tail())

print("\n" + "="*80)
print(" MODEL HAZIRLIĞI - GELİŞMİŞ")
print("="*80)

# En yoğun lokasyon
top_geohash = df['GEOHASH'].value_counts().index[0]
df_single = df[df['GEOHASH'] == top_geohash].copy().sort_values('DATE_TIME').reset_index(drop=True)

print(f"   Seçilen: {top_geohash}")
print(f"   Veri: {len(df_single)} satır")

# One-hot encoding
df_single = pd.get_dummies(df_single, columns=['Zaman_Dilimi'], drop_first=True)

target_col = 'AVERAGE_SPEED'
feature_cols = [
    'Saat', 'Haftanin_Gunu', 'HaftaSonu', 'Ay', 'DakikaGunun',
    'NUMBER_OF_VEHICLES', 'Lokasyon_Ort_Hiz', 'Lokasyon_Std_Hiz', 'Tatil_Gunu',
    'Saat_Sin', 'Saat_Cos', 'Ay_Sin', 'Ay_Cos', 'Gun_Hafta_Sin', 'Gun_Hafta_Cos',
    'Sinyal_Yogunlugu'
]
feature_cols.extend([col for col in df_single.columns if 'Zaman_Dilimi_' in col])

print("\n Kapsamlı lag features oluşturuluyor...")
for lag in [1, 2, 3, 4, 6, 8, 12, 16, 24, 48, 72]:
    df_single[f'speed_lag_{lag}'] = df_single[target_col].shift(lag)
    feature_cols.append(f'speed_lag_{lag}')

for window in [2, 3, 4, 6, 8, 12, 24]:
    df_single[f'speed_rolling_mean_{window}'] = df_single[target_col].rolling(window).mean()
    df_single[f'speed_rolling_std_{window}'] = df_single[target_col].rolling(window).std()
    feature_cols.append(f'speed_rolling_mean_{window}')
    feature_cols.append(f'speed_rolling_std_{window}')

for span in [3, 6, 12, 24]:
    df_single[f'speed_ewm_{span}'] = df_single[target_col].ewm(span=span).mean()
    feature_cols.append(f'speed_ewm_{span}')

# Hız değişim oranı
df_single['speed_change'] = df_single[target_col].diff()
df_single['speed_change_pct'] = df_single[target_col].pct_change()
feature_cols.extend(['speed_change', 'speed_change_pct'])

for lag in [1, 2, 4, 8]:
    df_single[f'vehicle_lag_{lag}'] = df_single['NUMBER_OF_VEHICLES'].shift(lag)
    feature_cols.append(f'vehicle_lag_{lag}')

for lag in [1, 2, 4, 8, 24]:
    df_single[f'sinyal_lag_{lag}'] = df_single['Sinyal_Yogunlugu'].shift(lag)
    feature_cols.append(f'sinyal_lag_{lag}')

df_single.dropna(inplace=True)

print(f"   ✓ Toplam özellik: {len(feature_cols)}")
print(f"   ✓ Temiz veri: {len(df_single)} satır")


print("\n Train/Test split...")
train_size = int(len(df_single) * 0.85)
train_data = df_single[:train_size]
test_data = df_single[train_size:]

print(f"   Train: {len(train_data)} | Test: {len(test_data)}")


scaler_X = RobustScaler()
scaler_y = MinMaxScaler()

X_train = scaler_X.fit_transform(train_data[feature_cols])
y_train = scaler_y.fit_transform(train_data[[target_col]])
X_test = scaler_X.transform(test_data[feature_cols])
y_test = scaler_y.transform(test_data[[target_col]])


def create_sequences(X, y, time_steps=24):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

TIME_STEPS = 24
X_train_seq, y_train_seq = create_sequences(X_train, y_train, TIME_STEPS)
X_test_seq, y_test_seq = create_sequences(X_test, y_test, TIME_STEPS)

print(f"   ✓ Train sequences: {X_train_seq.shape}")
print(f"   ✓ Test sequences: {X_test_seq.shape}")





np.random.seed(42)
tf.random.set_seed(42)

model = Sequential([

    Bidirectional(LSTM(
        256, return_sequences=True, activation='tanh',
        recurrent_dropout=0.2,
        kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)
    ), input_shape=(TIME_STEPS, X_train_seq.shape[2])),
    BatchNormalization(),
    Dropout(0.3),


    Bidirectional(GRU(
        192, return_sequences=True, activation='tanh',
        recurrent_dropout=0.2,
        kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)
    )),
    BatchNormalization(),
    Dropout(0.3),


    LSTM(128, return_sequences=True, activation='tanh',
         recurrent_dropout=0.15),
    BatchNormalization(),
    Dropout(0.25),


    LSTM(96, return_sequences=False, activation='tanh'),
    BatchNormalization(),
    Dropout(0.25),


    Dense(64, activation='relu', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)),
    BatchNormalization(),
    Dropout(0.2),

    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.15),

    Dense(16, activation='relu'),
    Dropout(0.1),

    Dense(1)
])


model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=0.001, weight_decay=0.0001),
    loss='huber',
    metrics=['mae', 'mse']
)

print(model.summary())

print("\n Model eğitimi başlıyor...")
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=25,
    restore_best_weights=True,
    verbose=1,
    min_delta=0.0001
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=0.00001,
    verbose=1
)
checkpoint = ModelCheckpoint(
    BASE_PATH + 'best_model_optimized.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=0
)

history = model.fit(
    X_train_seq, y_train_seq,
    epochs=200,
    batch_size=32,
    validation_split=0.15,
    callbacks=[early_stop, reduce_lr, checkpoint],
    verbose=1
)

print(f"\n Eğitim tamamlandı!")


print("\n" + "="*80)
print(" MODEL DEĞERLENDİRME")
print("="*80)

y_train_pred = model.predict(X_train_seq, verbose=0)
y_test_pred = model.predict(X_test_seq, verbose=0)

y_train_actual = scaler_y.inverse_transform(y_train_seq)
y_train_pred_scaled = scaler_y.inverse_transform(y_train_pred)
y_test_actual = scaler_y.inverse_transform(y_test_seq)
y_test_pred_scaled = scaler_y.inverse_transform(y_test_pred)


train_rmse = np.sqrt(mean_squared_error(y_train_actual, y_train_pred_scaled))
train_mae = mean_absolute_error(y_train_actual, y_train_pred_scaled)
train_r2 = r2_score(y_train_actual, y_train_pred_scaled)

test_rmse = np.sqrt(mean_squared_error(y_test_actual, y_test_pred_scaled))
test_mae = mean_absolute_error(y_test_actual, y_test_pred_scaled)
test_r2 = r2_score(y_test_actual, y_test_pred_scaled)
test_mape = np.mean(np.abs((y_test_actual - y_test_pred_scaled) / (y_test_actual + 1e-8))) * 100

print("\n" + "="*80)
print(" "*25 + " FINAL SONUÇLAR")
print("="*80)
print(f"\n🔹 TRAIN:")
print(f"   RMSE: {train_rmse:6.2f} km/h")
print(f"   MAE:  {train_mae:6.2f} km/h")
print(f"   R²:   {train_r2:7.4f} {'🏆' if train_r2 > 0.90 else ''}")
print(f"\n🔹 TEST:")
print(f"   RMSE: {test_rmse:6.2f} km/h")
print(f"   MAE:  {test_mae:6.2f} km/h")
print(f"   R²:   {test_r2:7.4f} {'🏆🏆🏆' if test_r2 > 0.85 else '🏆🏆' if test_r2 > 0.80 else '🏆'}")
print(f"   MAPE: {test_mape:6.2f}%")


print("\n" + "="*80)
print(" AKILLI ŞEHİR ANLIK MÜDAHALE ÖNERİLERİ")
print("="*80)

print("\n  VMS - Akıllı Rota Yönlendirme Simülasyonu")
print("-"*80)
ALARM_ESIGI = 25
kritik_indices = np.where(y_test_pred_scaled < ALARM_ESIGI)[0]

kritik_tahmin_bulundu = False
if len(kritik_indices) > 0:
    kritik_tahmin_bulundu = True
    idx = kritik_indices[0]
    kritik_zaman_index = test_data.index[idx + TIME_STEPS]
    kritik_zaman = df_single.loc[kritik_zaman_index]['DATE_TIME']
    kritik_hiz = y_test_pred_scaled[idx].item()

    print(f"  ALARM: {top_geohash} lokasyonunda kritik yavaşlama öngörülüyor!")
    print(f"   Zaman: {kritik_zaman}")
    print(f"   Tahmin Edilen Hız: {kritik_hiz:.1f} km/h")
    print(f"\n   📱 ÖNERİLEN VMS MESAJI:")
    print("   " + "="*60)
    print(f"   | DİKKAT: {top_geohash.upper()} BÖLGESİNDE YOĞUNLUK BEKLENİYOR!      |")
    print(f"   | LÜTFEN ALTERNATİF GÜZERGAHLARI KULLANIN.         |")
    print("   " + "="*60)
else:
    print(" Test döneminde VMS alarmı gerektirecek kritik bir yavaşlama öngörülmedi.")

print("\n\n [2] İETT - Toplu Taşıma Optimizasyon Simülasyonu")
print("-"*80)
otobus_guzergahlari = {
    "34G (Metrobüs)": [top_geohash, "sxk9w", kara_noktalar.iloc[0]['GEOHASH']],
    "500T (Tuzla - Cevizlibağ)": ["sxk8z", top_geohash, kara_noktalar.iloc[1]['GEOHASH']],
    "15F (Beykoz - Kadıköy)": ["sxk3m", top_geohash]
}

if kritik_tahmin_bulundu:
    etkilenen_hatlar = [hat for hat, guzergah in otobus_guzergahlari.items() if top_geohash in guzergah]

    if etkilenen_hatlar:
        gecikme = 10 + (30 - kritik_hiz) * 0.5
        print(f"  Etkilenen hatlar: {', '.join(etkilenen_hatlar)}")
        print(f"   Tahmini gecikme: {gecikme:.0f} dakika")
        print(f"\n    İETT OPERASYON MERKEZİ İÇİN PROAKTİF ÖNERİLER:")
        print(f"   • Yolcuları duraklardaki dijital ekranlarda '{gecikme:.0f} dk gecikme' uyarısıyla bilgilendirin.")
        print(f"   • Yoğunluk beklenen {etkilenen_hatlar[0]} hattına ek sefer planlamayı değerlendirin.")
else:
    print(" Toplu taşımayı etkileyecek kritik bir yavaşlama öngörülmedi.")

# Görselleştirme
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(f'LSTM Model - Test R²={test_r2:.4f}', fontsize=18, fontweight='bold')

# Loss
axes[0,0].plot(history.history['loss'], label='Train', linewidth=2)
axes[0,0].plot(history.history['val_loss'], label='Validation', linewidth=2)
axes[0,0].set_title('Loss', fontweight='bold')
axes[0,0].set_ylabel('Huber Loss')
axes[0,0].set_xlabel('Epoch')
axes[0,0].legend()
axes[0,0].grid(alpha=0.3)

# MAE
axes[0,1].plot(history.history['mae'], label='Train', linewidth=2)
axes[0,1].plot(history.history['val_mae'], label='Validation', linewidth=2)
axes[0,1].set_title('MAE', fontweight='bold')
axes[0,1].set_ylabel('MAE (km/h)')
axes[0,1].set_xlabel('Epoch')
axes[0,1].legend()
axes[0,1].grid(alpha=0.3)

# Predictions
sample = min(200, len(y_test_actual))
axes[0,2].plot(y_test_actual[:sample], label='Gerçek', linewidth=2, alpha=0.8)
axes[0,2].plot(y_test_pred_scaled[:sample], label='Tahmin', linewidth=2, alpha=0.8)
axes[0,2].set_title('Test: Gerçek vs Tahmin', fontweight='bold')
axes[0,2].set_ylabel('Hız (km/h)')
axes[0,2].legend()
axes[0,2].grid(alpha=0.3)

# Scatter
axes[1,0].scatter(y_test_actual, y_test_pred_scaled, alpha=0.4, s=15)
axes[1,0].plot([y_test_actual.min(), y_test_actual.max()],
               [y_test_actual.min(), y_test_actual.max()],
               'r--', linewidth=2)
axes[1,0].set_title(f'Scatter (R²={test_r2:.4f})', fontweight='bold')
axes[1,0].set_xlabel('Gerçek')
axes[1,0].set_ylabel('Tahmin')
axes[1,0].grid(alpha=0.3)

# Error dist
errors = y_test_actual.flatten() - y_test_pred_scaled.flatten()
axes[1,1].hist(errors, bins=60, color='coral', edgecolor='black', alpha=0.7)
axes[1,1].axvline(0, color='red', linestyle='--', linewidth=2)
axes[1,1].set_title('Hata Dağılımı', fontweight='bold')
axes[1,1].set_xlabel('Hata (km/h)')
axes[1,1].grid(alpha=0.3)

# Residual
axes[1,2].scatter(y_test_pred_scaled, errors, alpha=0.4, s=15)
axes[1,2].axhline(0, color='red', linestyle='--', linewidth=2)
axes[1,2].set_title('Residual Plot', fontweight='bold')
axes[1,2].set_xlabel('Tahmin')
axes[1,2].set_ylabel('Hata')
axes[1,2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig(BASE_PATH + 'optimized_results.png', dpi=300, bbox_inches='tight')
plt.show()

model.save(BASE_PATH + 'traffic.h5')




hiz_tahmini = ctrl.Antecedent(np.arange(0, 101, 1), 'hiz_tahmini')
arac_sayisi = ctrl.Antecedent(np.arange(0, df['NUMBER_OF_VEHICLES'].max() + 1, 1), 'arac_sayisi')
yesil_sure = ctrl.Consequent(np.arange(-30, 31, 1), 'yesil_sure')

hiz_tahmini['yavas'] = fuzz.trimf(hiz_tahmini.universe, [0, 0, 35])
hiz_tahmini['orta'] = fuzz.trimf(hiz_tahmini.universe, [25, 50, 75])
hiz_tahmini['hizli'] = fuzz.trimf(hiz_tahmini.universe, [65, 100, 100])

arac_max = df['NUMBER_OF_VEHICLES'].max()
arac_sayisi['az'] = fuzz.trimf(arac_sayisi.universe, [0, 0, arac_max * 0.4])
arac_sayisi['orta'] = fuzz.trimf(arac_sayisi.universe, [arac_max * 0.3, arac_max * 0.6, arac_max * 0.9])
arac_sayisi['cok'] = fuzz.trimf(arac_sayisi.universe, [arac_max * 0.8, arac_max, arac_max])

yesil_sure['azalt'] = fuzz.trimf(yesil_sure.universe, [-30, -20, -10])
yesil_sure['normal'] = fuzz.trimf(yesil_sure.universe, [-10, 0, 10])
yesil_sure['arttir'] = fuzz.trimf(yesil_sure.universe, [10, 20, 30])

# Kurallar
rule1 = ctrl.Rule(hiz_tahmini['yavas'] & arac_sayisi['cok'], yesil_sure['arttir'])
rule2 = ctrl.Rule(hiz_tahmini['orta'] & arac_sayisi['orta'], yesil_sure['normal'])
rule3 = ctrl.Rule(hiz_tahmini['hizli'] & arac_sayisi['az'], yesil_sure['azalt'])
rule4 = ctrl.Rule(hiz_tahmini['yavas'] & arac_sayisi['orta'], yesil_sure['arttir'])
rule5 = ctrl.Rule(hiz_tahmini['hizli'] & arac_sayisi['cok'], yesil_sure['azalt'])

control_sys = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
trafik_sim = ctrl.ControlSystemSimulation(control_sys)

# Test
if len(y_test_pred_scaled) > 0:
    idx = np.random.randint(0, len(y_test_pred_scaled))
    test_hiz = y_test_pred_scaled[idx].item()
    test_arac = test_data.iloc[idx + TIME_STEPS]['NUMBER_OF_VEHICLES']

    trafik_sim.input['hiz_tahmini'] = test_hiz
    trafik_sim.input['arac_sayisi'] = test_arac

    try:
        trafik_sim.compute()
        sonuc = trafik_sim.output.get('yesil_sure', 0.0)

        print(f"\n Fuzzy Logic Analizi:")
        print(f"   Tahmin Hız: {test_hiz:.1f} km/h")
        print(f"   Araç: {test_arac:.0f}")
        print(f"   → Yeşil Süre Değişimi: {sonuc:+.1f}%")

        if sonuc > 10:
            print(f"   🟢 Yeşil süre {sonuc:.0f}% ARTIRILMALI")
        elif sonuc < -10:
            print(f"   🔴 Yeşil süre {abs(sonuc):.0f}% AZALTILMALI")
        else:
            print(f"   🟡 Mevcut süre KORUNMALI")
    except:
        print("   ℹ️  Fuzzy hesaplama atlandı (kurallar tetiklenmedi)")



print(f"\n MODEL SONUÇ:")
print(f"   • Test R²:      {test_r2:7.4f} {'MÜKEMMEL' if test_r2 > 0.85 else 'İYİ' if test_r2 > 0.75 else '✓'}")
print(f"   • Test RMSE:    {test_rmse:6.2f} km/h")
print(f"   • Test MAE:     {test_mae:6.2f} km/h")
print(f"   • Test MAPE:    {test_mape:6.2f}%")


```
